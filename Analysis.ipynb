## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Loading Dataset

```python
df = pd.read_csv('hotel_booking.csv')
```

```python
df = df[df['adr'] < 5000]
```

## Data Analysis and Visualizations

```python
cancelled_pct = df['is_canceled'].value_counts(normalize = True)
print(cancelled_pct*100)

plt.figure(figsize = (4,4))
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), width=0.6)
plt.title('RESERVATION STATUS COUNT')
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x= df['hotel'], hue= df['is_canceled'], width=0.6, palette= 'Blues')
plt.title('Reservation Status in different hotels')
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.plot()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date').adr.mean()
city_hotel = city_hotel.groupby('reservation_status_date').adr.mean()
```

```python
resort_hotel
```

```python
plt.figure(figsize= (16,8))
plt.plot(resort_hotel.index, resort_hotel, label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel, label='City Hotel')
plt.title('Average daily rate in different hotel')
plt.legend(fontsize= 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(16,8))
ax1= sns.countplot(x= df['month'], hue= df['is_canceled'])
plt.title('Reservation status per month')
plt.xlabel('Months')
plt.ylabel('Number of reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
plt.figure(figsize=(16,8))
ax1= sns.barplot(x= 'month', y= 'adr', data= df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.title('ADR per month')
plt.xlabel('Months')
plt.ylabel('Average ADR value by months')
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]

top_10_countries = cancelled_data['country'].value_counts().head(10)
plt.figure(figsize = (8,8))
plt.pie(top_10_countries, autopct= '%.2f', labels= top_10_countries.index)
plt.title('Top 10 countries with reservations cancelled', fontsize= 15)
plt.show()
```

```python
df['market_segment'].value_counts(normalize= True)
```

```python
cancelled_data['market_segment'].value_counts(normalize= True)
```

```python
not_cancelled_data = df[df['is_canceled'] == 0]
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()

not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()
```

```python
plt.figure(figsize= (20,6))
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.title('Average Daily Rate', fontsize= 30)
plt.legend()
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]

not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize= (20,6))
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.title('Average Daily Rate', fontsize= 30)
plt.legend()
plt.show()
```
